## Regular Expression

Regular expression is a sequence of characters which forms a pattern.

The pattern can be used for searching, matching and manipulation.

Eg: Googel search, online registration matching etc

### Basic concepts

* Literal ch: Making the exact match. Eg: 'the' with 'the'.

* Meta ch: Ch with special meaning. Eg: '.'=> Any ch.

* Quantifier: Specific occurrences. Eg: '*'=> 0 or more occurrences,

                                         '+'=> 1 or more occurrences.

* Anchors: Define position.  Eg: '^'=> Starting, '$'=> Ending.

* Ch classes:Allows group matching. Eg:[A-Z],[A-Za-z0-9]

```python
theme="""Today is Anish's birthday. His friend Avi came to know about this one week back on
10/09/2025. They wanted to have a party and Avi decided to gift a I-phone 17 costing Rs.150000.
They exchanged contact details for easy coordination. 
Anish:+91-12345-67890, anish.joshi@nmims.edu.in; 
Avi:+91-98765-43210, avi_jain@gmail.com. They decided to meet on 20-09-2025"""
```

```python
print(theme)
```

```python
import re
```

```python
# Finding specific word
regx=re.findall(r'Anish',theme)
print(regx)
```

```python
# Finding specific word
regx=re.findall(r'Anish.',theme)
print(regx)
```

```python
# Finding specific word
regx=re.findall(r'Anish..',theme)
print(regx)
```

```python
# Finding specific word
regx=re.findall(r'..Anish...',theme)
print(regx)
```

```python
# Finding capitalized words
regx=re.findall(r'[A-Z][a-z]*',theme)
print(regx)
```

```python
# Finding capitalized words
regx=re.findall(r'[A-Z][a-z]+',theme)
print(regx)
```

```python
# Finding Rs.
regx=re.findall(r'Rs.[0-9]+',theme)
print(regx)
```

```python
# Finding Rs.
regx=re.findall(r'Rs.\d{6}',theme)
print(regx)
```

```python
theme_1="""Today is Anish's birthday. His friend Avi came to know about this one week back on
10/09/2025. They wanted to have a party and Avi decided to gift a I-phone 17 costing Rs.1,50,000.
They exchanged contact details for easy coordination. 
Anish:+91-12345-67890, anish.joshi@nmims.edu.in; 
Avi:+91-98765-43210, avi_jain@gmail.com. They decided to meet on 20-09-2025"""
```

```python
# Finding Rs.
regx=re.findall(r'Rs.\d{1},\d{2},\d{3}',theme_1)
print(regx)
```

```python
# Fetching date.
regx=re.findall(r'\d{2}/\d{2}/\d{4}',theme_1)
print(regx)
```

```python
# Fetching date.
regx=re.findall(r'\d{2}-\d{2}-\d{4}',theme_1)
print(regx)
```

```python
# Fetching date.
regx=re.findall(r'\d{2}[/-]\d{2}[-/]\d{4}',theme_1)
print(regx)
```

```python
# Fetching Phone nos
regx=re.findall(r'\+91-\d{5}-\d{5}',theme_1)
print(regx)
```

```python
# Fetching Email IDs
regx=re.findall(r'[A-Za-z0-9_.]+@[A-Za-z.]+',theme_1)
print(regx)
```
