## Handling null values in a dataframe

```python
df=pd.DataFrame([[10,20,30,np.nan,None],
                 [np.nan,np.nan,np.nan,40,50],
                 [np.nan,np.nan,np.nan,np.nan,np.nan],
                 [10,30,50,70,90]
                ])
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(how='any')
```

```python
df.dropna(how='all')
```

```python
# Dropping col

df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(axis=1,how='all')
```

```python
# Using threshold
df.dropna(thresh=1)
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=4)
```

## Filling values

Can be done: 1) Using value ; 2) Using method

```python
# Using value
df
```

```python
df.fillna(value=100)
```

```python
df
```

```python
df.fillna(100)
```

```python
# Filling columnwise

df.fillna({0:100,1:200,2:300,3:400,4:500})
```

```python
df
```

```python
# Fill row 1 by 50
df.loc[1].fillna(value=50)
```

```python
# Row 1,2 by 100

df.loc[[1,2]].fillna(100)
```

```python
# Using method

df
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
# Using statistics

df.fillna(value=df.mean())
```

```python
df.fillna(value=df.median())
```
