## Function which returns a list 

```python
# A fn which receives a no and returns a list of numbers

def returns_list(num):
    my_list=[]
    for val in range(num):
        my_list.append(val)
    return my_list
```

```python
print('The list of nos <10:',returns_list(10))
```

```python
print('The list of nos <20:',returns_list(20))
```

```python
# Update to get the list of even nos

def returns_even_list(num):
    even_list=[]
    for val in range(num):
        if val%2==0:
            even_list.append(val)
    return even_list
```

```python
print('The list of even nos <10:',returns_even_list(10))
```

```python
print('The list of even nos <20:',returns_even_list(20))
```

```python
# Update to get the list of odd nos

def returns_odd_list(num):
    odd_list=[]
    for val in range(num):
        if val%2!=0:
            odd_list.append(val)
    return odd_list
```

```python
print('The list of odd nos <10:',returns_odd_list(10))
```

```python
print('The list of odd nos <20:',returns_odd_list(20))
```

## Function which receives a list 

```python
def receives_list(my_list):
    my_sum=0
    for val in my_list:
        my_sum=my_sum+val
    avg=my_sum/len(my_list)
    return avg
```

```python
my_list=[1,3,5,7,9]
print('The average of the list:',receives_list(my_list))
```

## Function which returns two values


```python
def returns_two(my_list):
    my_sum=0
    for val in my_list:
        my_sum=my_sum+val
    avg=my_sum/len(my_list)
    return my_sum,avg
```

```python
print('The sum and average of the list elements:',returns_two(my_list))
```

```python
print('The sum of the list elements:',returns_two(my_list)[0])
print('The average of the list elements:',returns_two(my_list)[1])
```

```python
# Update the fn to return sum, length and avg of the list elements

def returns_three_values(my_list):
    my_sum=0
    for val in my_list:
        my_sum=my_sum+val
    avg=my_sum/len(my_list)
    return my_sum,len(my_list),avg
```

```python
result=returns_three_values(my_list)
print('The sum of the list elements:',result[0])
print('The length of the list:',result[1])
print('The average of the list elements:',result[2])

```

```python
# Define a fn which receives a list and return:
# 1) A list of squares of elements of the list
# 2) A list of cubes of elements of the list
# 3) Sum of the numbers in square list
# 4) Average of the cube list elements

def return_many(my_list):
    sq_list=[]
    cube_list=[]
    for val in my_list:
        sq_list.append(val**2)
        cube_list.append(val**3)
    sq_list_sum=sum(sq_list)
    cube_list_avg=sum(cube_list)/len(cube_list)
    return sq_list,cube_list,sq_list_sum,cube_list_avg
```

```python
result=return_many(my_list)
print(result)
```

```python
print('The list of squares:', result[0])
print('The list of cubes:',result[1])
print('The sum of sq list:',result[2])
print('The average of cube list:',result[3])
```
