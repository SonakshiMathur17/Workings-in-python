## Creation of a Car class

```python
class Car:
    def __init__(self,brand,model,year,color,price):
        self.brand=brand
        self.model=model
        self.year=year
        self.color=color
        self.price=price

    def display_features(self):
        print('The brand of the car is:',self.brand)
        print('The model of the car is:',self.model)
        print('The year of the car is:',self.year)
        print('The color of the car is:',self.color)
        print('The price of the car is:',self.price)
        
```

## Create a Marklist class

```python
class Marklist:
    def __init__(self,ica1_list,ica2_list,tee_list):
        self.ica1_list=ica1_list
        self.ica2_list=ica2_list
        self.tee_list=tee_list

    def ica1_avg(self):
        avg1=sum(self.ica1_list)/len(self.ica1_list)
        print('The ICA - 1 average is:', avg1)

    def ica2_avg(self):
        avg2=sum(self.ica2_list)/len(self.ica2_list)
        print('The ICA - 2 average is:', avg2)

    def tee_avg(self):
        avg3=sum(self.tee_list)/len(self.tee_list)
        print('The TEE average is:', avg3)
```

```python
# Creating an object

m1=Marklist([12,14,10,11,13,14],
            [13,14,15,11,10,9],
            [16,18,14,19,15,20])
```

```python
m1.ica1_list
```

```python
m1.ica2_list
```

```python
m1.tee_list
```

```python
m1.ica1_avg()
```

```python
m1.ica2_avg()
```

```python
m1.tee_avg()
```

## Inheritance

```python
class Parent: # super class
    def __init__(self,surname,wealth):
        self.name=surname
        self.wealth=wealth

    def display_name(self):
        print('The surname is :',self.name)

    def display_wealth(self):
        print('The wealth is :',self.wealth)
```

```python
# Creating a sub class

class Child(Parent):
    pass
```

```python
# Creating an object of the sub class

c1=Child()
```

```python
c1=Child('Jain',3000000)
```

```python
c1.name
```

```python
c1.wealth
```

```python
c1.display_name()
```

```python
c1.display_wealth()
```

## Sub class with attribute

```python
class Kid(Parent):
    def __init__(self,earned):
        self.earned=earned
    def display_earned(self):
        print('The earned wealth is:',self.earned)
```

```python
# Creating an object

k1=Kid()
```

## Using super()

```python
class Kid(Parent):
    def __init__(self,surname,wealth,earned):
        super().__init__(surname,wealth)
        self.earned=earned
    def display_earned(self):
        print('The earned wealth is:',self.earned)
```

```python
k2=Kid('Jain',5000000,25000)
```

```python
k2.name
```

```python
k2.wealth
```

```python
k2.earned
```

```python
k2.display_name()
```

```python
k2.display_wealth()
```

```python
k2.display_earned()
```
