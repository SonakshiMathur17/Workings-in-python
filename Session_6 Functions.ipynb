```python
# Define a fucntion with a string argument
# and display the no of characters in that string.

def len_string(string):
    print('The string is:',string)
    print('The length of the string is:', len(string))
```

```python
# Calling
len_string('Today is Wednesday and thank God, tomorrow is a holiday')
```

```python
# Update the fn to provide the words in the string

def update_len_string(string):
    print('The string is:',string)
    print('The length of the string is:', len(string))
    print('The list of words in the string:',string.split())
```

## Function which returns a value

```python
def add(n1,n2):
    summation=n1+n2
    return summation
```

```python
num1=int(input('Supply a no:'))
num2=int(input('Supply another no:'))
print('Sum of the nos:',add(num1,num2))
```

```python
# Define functions: add(),sub(), prod(), div(),quot(),
# and rem(), each with 2 args. 

def add(n1,n2):
    return(n1+n2)

def sub(n1,n2):
    return(n1-n2)

def prod(n1,n2):
    return(n1*n2)

def div(n1,n2):
    return(n1/n2)

def quot(n1,n2):
    return(n1//n2)

def rem(n1,n2):
    return(n1%n2)
```

```python
num1=int(input('Supply a no:'))
num2=int(input('Supply another no:'))
print('The sum:',add(num1,num2))
print('The sub:',sub(num1,num2))
print('The prod:',prod(num1,num2))
print('The div:',div(num1,num2))
print('The quot:',quot(num1,num2))
print('The ramainder:',rem(num1,num2))
```

```python
string1="Today is Wednesday and thank God tomorrow is a holiday"
```

```python
string2= "Today is Wednesday and tomorrow is Thursday"
```

```python
def common_words(s1,s2):
    print('The first string:',s1)
    s1_words=s1.split()
    print('The list of words in s1:',s1_words)
    print('The second string:',s2)
    s2_words=s2.split()
    print('The list of words in s2:',s2_words)

    common_list=[]
    for word in s1_words:
        if word in s2_words:
            common_list.append(word)
    print('The common words:', common_list)
            
```

```python
common_words(string1, string2)
```
