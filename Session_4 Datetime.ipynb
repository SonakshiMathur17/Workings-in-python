## Creating a pandas Series with datetime

```python
dates=[datetime(2025,9,26),
       datetime(2025,9,25),
       datetime(2025,9,24),
       datetime(2025,9,23),
       datetime(2025,9,22),
       datetime(2025,9,21)]
dates
```

```python
type(dates)
```

```python
type(dates[0])
```

```python
dates[0]
```

```python
dates[0:3]
```

```python
dates[3:]
```

```python
np.random.seed(100)
ts=pd.Series(np.random.rand(6),index=dates)
ts
```

```python
ts.loc['2025-09-26']
```

```python
ts.loc['2025-09-24':]
```

```python
ts.iloc[2:] 
```

### Creating a line graph


```python
plt.plot(ts);
```

```python
ts.max()
```

```python
# Provide the day which returned the max value
ts.idxmax()
```

```python
ts.idxmax().day
```

```python
# Average value of the series
ts.mean()
```

```python
# Find the date which gives the minimum value
ts.idxmin()
```

## Creating a pandas dataframe with datetime index

```python
dates=pd.date_range('2025-09-01',periods=26)
dates
```

```python
np.random.seed(100)
df=pd.DataFrame(np.random.randint(1000,size=(26,4)),index=dates)
df
```

```python
# Col 1

df[1]
```

```python
df[[1,2]]
```

```python
# Row

df.loc['2025-09-10']
```

```python
# Max of Col 2
df[2].max()
```

```python
# Provide the date which returned max Col 2
df[2].idxmax()
```

```python
# Mean of 15th
df.loc['2025-09-15'].mean()
```

```python
# Min of col 3 after 15th 

df.loc['2025-09-15':][3].min()
```

```python
# Which day provided the col 1 value max before 15th

df.loc[:'2025-09-15'][1].idxmax()
```

### Creating a loine graph

```python
plt.plot(df);
```

```python
plt.plot(df[0],c='r',label='Col 0')
plt.plot(df[1],c='b',label='Col 1')
plt.plot(df[2],c='g',label='Col 2')
plt.plot(df[3],c='y',label='Col 3')
plt.xlabel(' Dates')
plt.ylabel(' Values')
plt.title(' Time Series....')
plt.legend()
plt.xticks(rotation=45);
```

```python
df.plot(subplots=True)
plt.show()
```
